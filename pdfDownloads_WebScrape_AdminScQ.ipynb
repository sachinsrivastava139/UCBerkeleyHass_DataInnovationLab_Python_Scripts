```python
#Import as dataframe
df = pd.read_csv('AdminScQ_urls.csv')
# Convert the dataframe to python list
links = df['URL'].to_list()
links
```

```python
# Opening the URL - YOU NEED TO PUT INSTITUTIONAL LOGIN HERE 
driver_path = os.path.join(os.getcwd(),'chromedriver')
driver1 = uc.Chrome()
driver1.get('https://journals.sagepub.com/action/ssostart')
```

```python
# Get all the links that are needed to be downloaded. It parses every page of the each issue (3 second delay added for loading time)
href_values = []
for z in links:
    driver1.get(z)
    time.sleep(3)
    page_html = driver1.page_source
    soup = BeautifulSoup(page_html,'html.parser')
    # Getting all the download links
    download_link_extractor = soup.find_all('a', attrs={'title':'download'})
    
    #all the href tag values (links) are extracted and then appended with ?download=true string
    for link in download_link_extractor:
        href = link.get('href')
        href_values.append(href+'?download=true')

# adding the base url to each link
href_values = ['https://journals.sagepub.com' + href_value for href_value in href_values]
href_values_modified = [url.replace('reader', 'pdf') for url in href_values]

#saving the dataframe
df = pd.DataFrame({'Links': href_values_modified})
df.to_csv("Final_Pages_to_Parse_AdminScQ.csv")
```
