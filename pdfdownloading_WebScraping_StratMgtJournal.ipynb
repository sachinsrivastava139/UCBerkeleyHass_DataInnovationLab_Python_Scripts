```python
#Import as dataframe
df = pd.read_csv('StratMgtJournal_links.csv')
```

```python
# Convert the dataframe to python list
links = df['journal_link'].to_list()
```

```python
# Opening the URL 
# url = 'https://onlinelibrary.wiley.com/toc/10970266/1980/1/1'
# driver_path = os.path.join(os.getcwd(),'chromedriver')
driver = uc.Chrome()
# driver.maximize_window()

# Looping through all pdf
url_list = []
for j in links:
    driver.get(j)
    # Download the html code
    time.sleep(5)
    page_html = driver.page_source
    soup = BeautifulSoup(page_html,'html.parser')
    # Getting all the download links
    download_link_extractor = soup.find_all('a', attrs={'title':'EPDF'})
    
    for i in download_link_extractor:
        url_list.append(i['href']+'?download=true')
        

driver.close()
print("done")
```

```python
#preparing the final download links and saving in a dataframe
test_string = " ".join(url_list)
modified_string = test_string.replace('epdf', 'pdfdirect')
modified_url_list = modified_string.split()
df = pd.DataFrame({'Links': modified_url_list})
df.to_csv("Final_Download_links.csv")
```
