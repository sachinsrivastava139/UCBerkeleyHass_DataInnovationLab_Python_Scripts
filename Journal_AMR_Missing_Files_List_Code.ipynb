```python
import os
import csv

# Set the path of the folder you want to read
folder_path = '/Users/sachinsrivastava/Downloads/PythonCodes/AMR_Downloads/AMR_v1.0'

# Get a list of all the file names in the folder
file_names = os.listdir(folder_path)

# Create an empty list to store the data
data = []

# Loop through each file name and append it to the data list
for file_name in file_names:
    # Remove the .pdf extension from the file name
    file_name = file_name[:-4]
    
    # Split the file name by the underscore delimiter
    split_file_name = file_name.split('_')
    
    # Add the split file name to the data list
    data.append(split_file_name)

# Define the column names for the CSV file
column_names = ['file name', 'column 1', 'column 2', 'column 3']

# Set the name and path of the CSV file you want to create
csv_file_name = 'Downloaded_List.csv'
csv_file_path = os.path.join(folder_path, csv_file_name)

# Open the CSV file for writing
with open(csv_file_path, 'w', newline='') as csv_file:
    # Create a CSV writer object
    csv_writer = csv.writer(csv_file)
    
    # Write the column names to the first row of the CSV file
    csv_writer.writerow(column_names)
    
    # Loop through the data and write each row to the CSV file
    for row in data:
        csv_writer.writerow(row)
        
print(f"CSV file '{csv_file_name}' created successfully at '{csv_file_path}'.")
```
