```python
#Import as dataframe
df = pd.read_csv('OrgSc_urls.csv')
df
```

```python
# Convert the dataframe to python list
links = df['URL'].to_list()
links
```

```python
# Opening the URL 
driver = uc.Chrome()
url_list = []

counter = 0
for j in links:
    driver.get(j)
    # Download the html code
    time.sleep(4)
    page_html = driver.page_source
    soup = BeautifulSoup(page_html,'html.parser')
    time.sleep(1)
    # Getting all the download links & post appending downbload=true
    download_link_extractor = soup.find_all('a', attrs={'title':'PDF'})
    time.sleep(1)
    for i in download_link_extractor:
        url_list.append(i['href']+'?download=true')
        print(counter)
        print(i)
        counter = counter + 1

time.sleep(10)
driver.close()
print("done")
```

```python
#replace and pre-appending CMU specific URl.
url_list_modified = [url.replace('/doi/epdf/', 'https://pubsonline-informs-org.cmu.idm.oclc.org/doi/pdf/') for url in url_list]
df = pd.DataFrame({'Links': url_list_modified})
df.to_csv("Final_Download_links_OrgSc.csv")
```
