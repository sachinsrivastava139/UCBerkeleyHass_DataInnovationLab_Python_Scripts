```python
#Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

#import packages
import json
import os
import spacy
import pandas as pd
import numpy as np

# Set the directory path
directory = '/content/drive/MyDrive/UC-BerkeleyHass_DataInnovationLab_Sachin/mgtsc_data/'
directory_save = '/content/drive/MyDrive/UC-BerkeleyHass_DataInnovationLab_Sachin/saved_sheets/'

nlp = spacy.load('en_core_web_sm')
results = []

# Loop through all files in the directory
for filename in os.listdir(directory):
    # Check if the file is a JSON file
    if filename.endswith('.json'):
        # Open the JSON file
        with open(os.path.join(directory, filename)) as f:
            # Load the JSON data into a dictionary
            paper = json.load(f)
            print(paper.keys())
            # Loop through all the sections in the paper
            for section in paper:
                # Extract named entities from the section
                doc = nlp(paper[section])
                for ent in doc.ents:
                    if ent.label_ == 'GPE':
                        results.append([filename, section, ent.text])

# Convert the results to a Pandas DataFrame and save as CSV
df = pd.DataFrame(results, columns=['filename', 'section', 'entity'])
output_file = os.path.join(directory_save, 'Final_Test5.csv')
df.to_csv(output_file, index=False)

# Print the DataFrame
print(df)
```
