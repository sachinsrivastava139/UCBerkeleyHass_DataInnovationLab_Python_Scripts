```python
# declare global path
file_path = '/Users/sachinsrivastava/Downloads/PythonCodes/'
Initial_path = "/Users/sachinsrivastava/Downloads/PythonCodes/AMR_Downloads"
```

```python
#Import as dataframe
df = pd.read_csv('Final_Download_links_AMR.csv')
df
```

```python
# Convert the dataframe to python list 
links = df['Links'].to_list()
links_modified = [sub.replace('https://doi.org/10.5465/', '') for sub in links]
links_modified 
```

```python
#Declare Chrome Options
options = uc.ChromeOptions()
options.add_experimental_option('prefs', {
"download.default_directory": "/Users/sachinsrivastava/Downloads/PythonCodes/AMR_Downloads", #Change default directory for downloads
"download.prompt_for_download": False, #To auto download the file
"download.directory_upgrade": True,
"plugins.always_open_pdf_externally": True #It will not show PDF directly in chrome
})

# Opening the URL 
driver = uc.Chrome(options=options)
# url_list = []
url = 'https://cmu.primo.exlibrisgroup.com/discovery/login?vid=01CMU_INST:01CMU&lang=en'
# counter = 0
driver.get(url)
```

```python
counter = 0
for j in links_modified[:]:
    
    driver.find_element(By.XPATH, '//*[@id="searchBar"]').send_keys(j)
    driver.find_element(By.XPATH, '//*[@id="searchBar"]').send_keys(Keys.RETURN)
    #Sleep option
    time.sleep(4)
    
    try:
        #download_pdf_button = driver.find_element(By.XPATH, '//*[@id="SEARCH_RESULT_RECORDID_cdi_crossref_primary_10_5465_amr_1976_19369636"]/div[3]/div[2]/prm-search-result-availability-line/div/div[1]/div/a/span')
        download_pdf_button = driver.find_element(By.XPATH, "//span[contains(text(),'Download PDF')]")
        download_pdf_button.click()
        #Sleep option
        time.sleep(4)
        
    except:
        print(str(counter)+"_Download Button Missing")
        counter = counter+1
        #Clear the Search Box
        clear_search_button = driver.find_element(By.XPATH, '//*[@id="searchElementsWrapper"]/div/form/div/div/div[1]/prm-autocomplete/md-autocomplete-wrap/button/prm-icon/md-icon')
        clear_search_button.click()
        continue

    #Tab Switch from Tab-0 to Tab-1
    driver.switch_to.window(driver.window_handles[1])
    #Sleep option
    time.sleep(7)

    #Clicking the EBSCO Download Button
    #download_pdf_button = driver.find_element(By.XPATH, '//*[@id="downloadLink"]')
    download_pdf_button = driver.find_element(By.XPATH, "//a[contains(text(),'Download PDF')]")
    download_pdf_button.click()
    #Sleep option
    time.sleep(15)

    # Name Change for each Downloaded File
    filename = max([Initial_path + "/" + f for f in os.listdir(Initial_path)],key=os.path.getctime)
    shutil.move(filename,os.path.join(Initial_path, str(counter)+"_"+j+'.pdf'))

    #Close tab-2
    driver.close()
    #Tab Switch from Tab-0 to Tab-1
    driver.switch_to.window(driver.window_handles[0])
    
    #Clear the Search Box
    clear_search_button = driver.find_element(By.XPATH, '//*[@id="searchElementsWrapper"]/div/form/div/div/div[1]/prm-autocomplete/md-autocomplete-wrap/button/prm-icon/md-icon')
    clear_search_button.click()
    
    print(counter)
    counter = counter+1
```
