```python
# Convert the dataframe to python list
links = df['URL'].to_list()
links
```

```python
# Opening the URL 
driver = uc.Chrome()
url_list = []
url = 'https://journals.aom.org/toc/amr/1/1'
counter = 0
driver.get(url)
```

```python
# Download the html code
time.sleep(4)
page_html = driver.page_source
soup = BeautifulSoup(page_html,'html.parser')
time.sleep(1)
```

```python
soup
```

```python
# Getting all the download links
download_link_extractor = soup.find_all('a', attrs={'title':'PDF/EPUB'})
time.sleep(1)
for i in download_link_extractor:
    url_list.append(i['href']+'?download=true')
    print(counter)
    print(i)
    counter = counter + 1
```

```python
# Opening the URL 
driver = uc.Chrome()
url_list = []

counter = 0
for j in links:
    driver.get(j)
    # Download the html code
    time.sleep(4)
    page_html = driver.page_source
    soup = BeautifulSoup(page_html,'html.parser')
    time.sleep(1)
    # Getting all the download links
    download_link_extractor = soup.find_all('a', attrs={'title':'PDF/EPUB'})
    time.sleep(1)
    for i in download_link_extractor:
        url_list.append(i['href'])
        print(counter)
        print(i)
        counter = counter + 1

time.sleep(10)
driver.close()
print(url_list)
print("done")
```
